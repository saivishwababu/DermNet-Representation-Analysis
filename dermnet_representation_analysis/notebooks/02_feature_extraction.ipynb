Extract embeddings from DINOv2, ResNet50, and CLIP models.

```python
import sys, os
import numpy as np
sys.path.insert(0, '..')

from src.utils import load_config, setup_logging, get_device
from src.data_loader import load_dataset
from src.preprocessing import remove_duplicates
from src.feature_extractors import DINOv2Extractor, ResNet50Extractor, CLIPExtractor

config = load_config('../config.yaml')
logger = setup_logging()
device = get_device()
print(f'Device: {device}')

df, stats = load_dataset(config, '..')
image_paths = df['image_path'].tolist()
labels = df['label'].tolist()

features_dir = os.path.join('..', config['outputs']['features_dir'])

# Extract DINOv2
dinov2 = DINOv2Extractor(device=device)
dinov2_feats = dinov2.extract_and_save(image_paths, labels, features_dir)
print(f'DINOv2 features shape: {dinov2_feats.shape}')
```
