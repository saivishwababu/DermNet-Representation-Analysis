# 04 — Clustering and Metrics

Apply K-Means, Agglomerative, and HDBSCAN to PCA-reduced features and compute evaluation metrics.

```python
import sys, os
import numpy as np
sys.path.insert(0, '..')

from src.utils import load_config
from src.clustering import run_all_clustering
from src.evaluation import evaluate_all

config = load_config('../config.yaml')
features_dir = os.path.join('..', config['outputs']['features_dir'])
tables_dir = os.path.join('..', config['outputs']['tables_dir'])

labels = np.load(os.path.join(features_dir, 'labels.npy'))
n_classes = len(set(labels))

# Load PCA features
pca_feats = {}
cluster_results = {}
for model in ['dinov2', 'resnet50', 'clip']:
    pca_path = os.path.join(features_dir, f'{model}_pca.npy')
    if os.path.exists(pca_path):
        pca_feats[model] = np.load(pca_path)
        cluster_results[model] = run_all_clustering(pca_feats[model], n_classes, config)

results_df = evaluate_all(
    pca_feats, cluster_results, labels,
    os.path.join(tables_dir, 'clustering_results.csv')
)
results_df
```
